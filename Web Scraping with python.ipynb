```python
# importing all the necessery library
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
urls = 'https://www.skysports.com/cricket/teams/india/results'
```

```python
response = requests.get(urls)
```

```python
response.status_code
```

```python
print(response.text)
```

```python
html_page_data = BeautifulSoup(response.text,'html.parser')
```

```python
date = html_page_data.find_all('h3',{'class':'text-h5 box strap3 -interact -caps -center'})
date
```

```python
actual_date = [i.text for i in date]
actual_date
```

```python
matches_name = html_page_data.find_all('h3',{'class':'box text-h6 -center strap1 -interact -caps'})
matches_name
```

```python
matches_name = [i.text for i in matches_name]
matches_name
```

```python
time_lst = html_page_data.find_all('span')
time_lst
```

```python
'9:00'[0].isnumeric()
```

```python
time_lst[7].text[0]
```

```python
actual_time = []
for i in time_lst:
    
    try:
        if i.text[0].isnumeric():
            print(i.text)
            actual_time.append(i.text)
    except Exception as e:
        pass
actual_time
```

```python
teams_name_lst = html_page_data.find_all('h3',{'class':'text-h3'})
teams_name_lst
```

```python
int(len(teams_name_lst)/2)
```

```python
actual_teams_name_lst = [f'{teams_name_lst[i].text} vs {teams_name_lst[i+1].text}' for i in range(0,len(teams_name_lst),2)]
```

```python
actual_teams_name_lst
```

```python
len(actual_teams_name_lst)
```

```python
dataframe = pd.DataFrame({'Date':actual_date,
                            'Series Name':matches_name,
                            'Time':actual_time,
                            'Matches Name':actual_teams_name_lst})
dataframe
```
